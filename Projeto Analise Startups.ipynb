```python
analise = round(dados['Pais'].value_counts(normalize = True) * 100,1)
```

```python
analise
```

```python
plt.figure(figsize = (15, 6))
plt.title('Analisando os Paises que mais geram unicornios no mundo')
plt.pie(
    analise,
    labels = analise.index,
    shadow = True, 
    startangle = 90, 
    autopct='%1.1f%%'
);
```

```python
plt.figure(figsize = (15, 6))
plt.title('Analisando top 10 Paises que mais geram empresas unicornios no mundo')
plt.pie(
    analise.head(10),
    labels = analise.index[0:10],
    shadow = True, 
    startangle = 90, 
    autopct='%1.1f%%');
```

```python
# Conversão para Data
dados['Data de inicio'] = pd.to_datetime(dados['Data de inicio'] )

dados['Data de inicio'].head()
```

```python
# extrair o ano e mes.

dados['mes'] = pd.DatetimeIndex(dados['Data de inicio']).month

dados['ano'] = pd.DatetimeIndex(dados['Data de inicio']).year

dados.head()
```

```python
#tabela analitica

tabela_analitica = dados.groupby( by = ['Pais','ano','mes']).count().reset_index()
tabela_analitica
```

```python
#filtrando por pais

tabela_analitica.loc[
    tabela_analitica['Pais'] == 'Brazil'
]
```

```python
# Transformando coluna do valor.

dados ['Valor ($)'] = pd.to_numeric(dados['Valor ($)'].apply(lambda linha:linha.replace('$','')))

dados.head()
```
